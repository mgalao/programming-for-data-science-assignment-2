**5.** Start by printing all the column values in the cell bellow.

**6.** List the values in the column 'Country Name'.You will get a list with repeated values, **delete all duplicates** to ease your analysis. <br/>

*Tip: There is a method in the pandas library that allows to do this easily.*

```python
# Print the unique country names
print(wdi["Country Name"].unique())
```

You might notice that while the bottom rows represent Countries, the top rows represent aggregates of countries (e.g., world regions). <br/> We are only interested in **working with country-level data**, and as such we need to filter out all the unnecessary rows.

**7.** Save all the values of column 'Country Name' in a variable called **cnames**. <br/>

```python
# Save the values of the "Country Name" column in the "cnames" variable
cnames = wdi["Country Name"]
```

**7.1.** Delete all duplicate values.<br>

```python
# Remove duplicate country names from the "cnames" Series
cnames.drop_duplicates(inplace=True)

# Display the unique country names
print(cnames.values)
```

**7.2.** Print the names that do not represent countries.

```python
# Manually analyze the "cnames" Series to determine which entries do not represent individual countries
# We found that all entries above "World" are aggregates (continents, regions, etc)

# Get the index of "World" in the "cnames" Series
world_index = cnames[cnames == "World"].index[0]

# Select all entries in "cnames" up to and including "World"
aggregates = cnames[cnames.index <= world_index]

# Print the names that do not represent countries
print(aggregates.values)
```

You can take advantage of the structure of the dataset to realize that aggregates (Continents, Regions, etc) are all located on the top of the series 'cnames'. Moreover, since the series is small you can easily validate this assumption manually and then use that information to extract a slice of all the entries that represent non-countries entities.<br/>

**8.** In the next cell filter out, from **wdi**, the rows in which 'Country Name' represents an aggregate of countries.<br/>

```python
# Filter the "wdi" DataFrame to exclude rows where "Country Name" is in the "aggregates" Series
wdi = wdi[~wdi['Country Name'].isin(aggregates)]
```

**9.** Check that the top of the **wdi** dataframe now only has countries and not aggregates of countries.

```python
# Display the unique country names (extra to confirm the operation)
print(wdi["Country Name"].unique())
```

## Indicator Codes and Indicator Names

**12.** Select the columns 'Indicator Name' and 'Indicator Code'.Then, delete all the duplicates, and print the top 5 and bottom 5 values. <br/>

**13.** Create a new DataFrame named **indicators** made up of the columns 'Indicator Name' and 'Indicator Code'. Then, delete all the duplicated entries. Finally, set the column 'Indicator Code' as the index of **indicators**. 

*Note: Try to perform all these steps in a single line of code.*

```python
# Create a new DataFrame "indicators" made up of the columns "Indicator Name" and "Indicator Code", delete all the duplicates, and set "Indicator Code" as the index
indicators = wdi[["Indicator Name", "Indicator Code"]].drop_duplicates().set_index("Indicator Code")

# Display the first 5 rows (to confirm the operation)
indicators.head()
```

**The 'indicators' DataFrame can operate now as a dictionary. <br/>**
By passing an 'Indicator Code' (key) it returns the associated 'Indicator Name' (value).<br/>

**14.** Using the **indicators** DataFrame, find the 'Indicator Code' associated with the following observables:
1. 'Population', find the 'Indicator Code' of the total population in a country;
2. 'GDP', find the GDP measured in current US Dollars;
3. 'GINI index'

*Hint: You can use the method STRING.str.contains('substring') to check whether a string contains a substring.*

```python
# Get the entries from the "indicators" DataFrame where the "Indicator Name" contains the substring "Population" (to help identify the correct total population indicator)
indicators[indicators['Indicator Name'].str.contains('Population')]

# The actual indicator name for total population is: "Population, total"
```

```python
# Get the entries from the "indicators" DataFrame where the "Indicator Name" contains the substrings "GDP" and "current", not case-sensitive (to help identify the correct GDP measured in current US dollars indicator)
indicators[indicators['Indicator Name'].str.contains('GDP', case=False) &
           indicators['Indicator Name'].str.contains('current', case=False)]

# The actual indicator name for GDP measured in current US dollars is: "GDP (current US$)"
```

```python
# Get the entries from the "indicators" DataFrame where the "Indicator Name" contains the substring "GINI", not case-sensitive (to help identify the correct GINI index indicator)
indicators[indicators['Indicator Name'].str.contains('GINI', case=False)]

# The actual indicator name for GINI index is: "GDP (current US$)"
```

```python
# Find the "Indicator Code" for total population
total_population_index = indicators[indicators['Indicator Name'] == 'Population, total'].index.tolist()[0]
print("'Indicator Code' for total population:", total_population_index)

# Find the "Indicator Code" for GDP measured in current US Dollars
gdp_current_us_dollar_index = indicators[indicators['Indicator Name'] == 'GDP (current US$)'].index.tolist()[0]
print("'Indicator Code' for GDP measured in current US Dollars:", gdp_current_us_dollar_index)

# Find the "Indicator Code" for the GINI index
gini_index = indicators[indicators['Indicator Name'].str.contains('GINI', case=False)].index.tolist()[0]
print("'Indicator Code' for GINI index:", gini_index)
```

## Extracting and Cleaning Data from WDI and PWT

**15.** From **wdi** extract the columns 'Indicator Code', 'Country Code', and '2012'.
Save the output in variable **wdi_sample**.

*Note: You should be able to perfom all operations in a single line of code. <br/>*

```python
# Extract the "Indicator Code", "Country Code", and "2012" columns from "wdi" DataFrame and save in "wdi_sample"
wdi_sample = wdi[["Indicator Code", "Country Code", "2012"]]

# Display the resulting DataFrame
wdi_sample
```

**16.** Select from **wdi_sample** the lines associated with the Indicator Codes that you found in question 14., which concern the data of the 'Population total','GDP', and 'GINI index'.

```python
# Select rows from "wdi_sample" based on the "Indicator Code" for total population, GDP measured in current US Dollars, and GINI index
wdi_sample = wdi_sample[wdi_sample['Indicator Code'].isin([total_population_index, gdp_current_us_dollar_index, gini_index])]

# Display the resulting filtered DataFrame
wdi_sample
```

**17.** Create a pivot table, in which the **values** are column '2012', the **index** is 'Country Code', and the **columns** are the Indicator Codes. <br/>

*Hint: Pandas has a very useful method to create pivot tables.*

```python
# Create a pivot table with "2012" as the values, "Country Code" as the index, and "Indicator Code" as the columns
wdi_sample = wdi_sample.pivot_table(values="2012", index="Country Code", columns="Indicator Code")

# Display the resulting pivot table
wdi_sample
```

**18.** Rename the column names of **wdi_sample** to 'Population', 'GDP', and 'GINI', accordingly.

```python
# Rename the columns of the pivot table
wdi_sample.rename(columns={total_population_index: "Population",
                           gdp_current_us_dollar_index: "GDP",
                           gini_index: "GINI"}, inplace=True)

# Display the updated pivot table with renamed columns
wdi_sample
```

**19.** From **pwt** select only the values of the year 2012. <br/>
Then, extract the columns 'countrycode' and 'hc' into a new variable **pwt_sample**. <br/>
Rename 'countrycode' to 'Country Code', so that it matches the same column in **wdi_sample**<br/>

*Note: in this case 'hc' stands for the Human Capital Index.<br/>*

```python
# Filter rows where the "year" is 2012, select the "countrycode" and "hc" columns, and rename "countrycode" to "Country Code"
pwt_sample = pwt.loc[pwt["year"] == 2012, ["countrycode", "hc"]].rename(columns={"countrycode": "Country Code"})

# Display the resulting DataFrame
pwt_sample
```

**20.** Finally, create a new dataframe named **data** that contains the columns from **wdi_sample** and **pwt_sample**, matched by 'Country Code'. 

*Hint: Use the method concat(), and make sure both dataframes have the same index ('Country Code').*

```python
# Concatenate "wdi_sample" with "pwt_sample" (after setting "Country Code" as the index) along columns (axis=1)
data = pd.concat([wdi_sample, pwt_sample.set_index("Country Code")], axis=1)

# Display the resulting concatenated DataFrame
data
```
