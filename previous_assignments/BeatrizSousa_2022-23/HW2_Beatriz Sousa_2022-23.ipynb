Perform the necessary manipulations to answer the following questions, unless otherwise stated you can use the country codes to represent the countries in your solutions:
1. Which countries have a **population size of 10 million habitations +/- 1 million**?
2. What is  the **average** and the **standard deviation in GDP** of countries listed in 1?
3. What is  the **average** and the **standard deviation in the GDP per capita** of countries listed in 1?
4. Consider the following classification of country size: <br/>
    Tiny - population < 1 000 000 <br/>
    Very Small - 1 000 000 <= population < 5 000 000 <br/>
    Small - 5 000 000 <= population < 15 000 000 <br/>
    Medium - 15 000 000 <= population < 30 000 000 <br/>
    Large - 30 000 000 <= population < 100 000 000 <br/>
    Huge - 100 000 000 <= population <br/>
   What is  the **average** and the **standard deviation in the GDP per capita of countries in each size classification**?   
5. Create a **function** that will take a dataframe and a column name, and **return** a series with binary values indicating whether the **values from the column are above the mean value of that column** (indicated with a value of 1 or 0 otherwise). If the value in the column is missing (NaN) the value in the series should also be missing (NaN). Test your function. *Hint:* search how to check if something is None so that we can return None. <br/>
6. What is the **average GDP per capita of the countries after being grouped by size classification and whether or not the human capital is above average**? 
7. What is the **average GDP per capita of countries after being grouped by whether or not the human capital is above average and whether or not the gini coefficient is above average?**
8. What is the **name of the country** that has the **highest GDP per capita + a Gini coefficient below average and a level of human capital below average**?
9. What is the **name of the country** that has the **highest GDP per capita + a Gini coefficient below average for its size classification and a level of human capital below average for its size classification**?
10. What is the **name of the country** that has the **largest % increase in GDP between 1980 and 2010?** *HINT: You will need to use the wdi dataframe.*

Print the answer at the end of that cell.

```python
#1.
# population_10_millions is a copy of a sliced dataframe, only with the countries that have population of 10 million +-1million
population_10_millions = data.loc[(data['Population'] >= 9000000) & (data['Population'] <= 11000000)].copy()
print('The countries have a population size of 10 million habitations +/- 1 million are', population_10_millions.index.tolist())
```

```python
#2.
# we use the functions that are alredy defined in pandas to calculate the average and the standard deviation of the column GDP
GDP_average = population_10_millions['GDP'].mean()
GDP_standard_deviation = population_10_millions['GDP'].std()
print('The average and the standard deviation in GDP of countries listed in 1 are respectively', GDP_average, 'and', GDP_standard_deviation)
```

```python
#3.
# we add another column to the population_10_millions dataframe with the GDP per capita (division between GDP and population)
population_10_millions['GDP per capita'] = population_10_millions['GDP'] / population_10_millions['Population']

# then we use the functions that are alredy defined in pandas to calculate the average and the standard deviation of the column GDP per capita
GDP_percapita_average = population_10_millions['GDP per capita'].mean()
GDP_percapita_standard_deviation = population_10_millions['GDP per capita'].std()
print('The average and the standard deviation in GDP per capita of countries listed in 1 are respectively', GDP_percapita_average, 'and', GDP_percapita_standard_deviation)
```

```python
#4.
# first we define a function to classify a country's size by its population
def size_classification(population):
    if population < 1000000:
        return 'Tiny'
    elif population < 5000000:
        return 'Very Small'
    elif population < 15000000:
        return 'Small'
    elif population < 30000000:
        return 'Medium'
    elif population < 100000000:
        return 'Large'
    elif population >= 100000000:
        return 'Huge'
    else:
        return 'No population data'
# then we add a column to the dataframe with this classification
data['Size Classification'] = data['Population'].apply(size_classification)

# we also add a column for the GDP per capita on the original dataframe, in the same way as before
data['GDP per capita'] = data['GDP'] / data['Population']

# finally we calculate the average and the standard deviation for each size classification, using the groupby() function
mean_by_size = data.groupby(['Size Classification'])[['GDP per capita']].mean()
standard_deviation_by_size = data.groupby(['Size Classification'])[['GDP per capita']].std()
print('The average of the GDP per capita by each size classification is\n', mean_by_size, '\n\nWhile the standard deviation of the GDP per capita by each size classification is\n', standard_deviation_by_size)
```

```python
#5.
def above_the_mean(data, column):
    # we start by calculating the average value of the column
    mean = data[column].mean()
    # then we define a name for the new column based on the one that we are comparing
    new_column_name = column + ' above mean'
    # and we create a binary column based on whether the value is above or below the average
    data[new_column_name] = (data[column] >= mean).astype(int)
    # finally we check if there are missing values in the original column, and if there are, we assign the new column also with a missing value
    data.loc[pd.isnull(data[column]), new_column_name] = np.NaN
    return data

# checking if it works for GDP
above_the_mean(data, 'GDP')
```

```python
#6.
# first we create a binary column for Human Capital Index
above_the_mean(data, 'Human Capital Index')

# then by size and by whether or not the HCI is above average we calculate the average of the GDP per capita of the countries grouped
mean_by_size_hci = data.groupby(['Size Classification', 'Human Capital Index above mean'])['GDP per capita'].mean()
print('The average of the GDP per capita by each size classification and HCI above the mean is\n', mean_by_size_hci)
```

```python
#7.
# first we create a binary column for Gini Index, since we already created one for the Human Capital Index
above_the_mean(data, 'GINI')

# then by whether or not the HCI is above average and by whether or not the Gini index is above average we calculate the average of the GDP per capita of the countries grouped
mean_by_hci_gini = data.groupby(['Human Capital Index above mean', 'GINI above mean'])['GDP per capita'].mean()

print('The average of the GDP per capita by whether or not HC is above average and whether or not the GINI is above average is\n', mean_by_hci_gini)
```

```python
#8.
# first we select the countries from data that satisfy the conditions requested, then we get the index of the maximum value of GDP per capita (with idxmax)
# we use iloc[0] to get only the country code
country_code = data.loc[(data['GINI above mean'] == 0) & (data['Human Capital Index above mean'] == 0), ['GDP per capita']].idxmax(axis = 0).iloc[0]
# then we search in the wdi dataframe for the name of the country that has the code we found before
country_name = wdi.loc[wdi['Country Code'] == country_code, ['Country Name']].iloc[0,0]

print('The country that has the highest GDP per capita, a Gini coefficient below average and a level of human capital below average is', country_name)
```

```python
#9.
# first we create a list with the different classifications that we find in the dataframe
sizes = data['Size Classification'].unique()

# we create two variables that store the maximum gdp per capita and the code of the country that has it
max_gdp = 0.0
max_country_code = ''

# we will run all sizes, and find the country that satisfies the requests for each size, and then we find the overall highest GDP per capita
for size in sizes:
    # we create a temporary dataframe, only with the countries in the size classification we are working on
    temp_data = data.loc[data['Size Classification'] == size, ['Size Classification', 'GDP', 'GINI', 'Population', 'Human Capital Index', 'GDP per capita']]
    # then we add the binary columns for both indicators, using the previous function
    above_the_mean(temp_data, 'GINI')
    above_the_mean(temp_data, 'Human Capital Index')
    # we filter the countries that satisfy both conditions (being below the average on both indicators)
    temp_conditions = temp_data.loc[(temp_data['GINI above mean'] == 0) & (temp_data['Human Capital Index above mean'] == 0)]
    # if we find countries that satisfy those conditions::::
    if not temp_conditions.empty:
        # we find the row associated with the maximum value of GDP per capita
        temp_max = temp_conditions[temp_conditions['GDP per capita'] == temp_conditions['GDP per capita'].max()]
        # and we compare the value with the maximum gdp we have until this moment (stored on max_gdp)
        if float(temp_max['GDP per capita'].iloc[0]) > max_gdp:
            # we save it if it's true
            max_gdp = temp_max['GDP per capita'].iloc[0]
            max_country_code = temp_max.index.tolist()[0]
            
# finally, having the country code we search for its name
max_country_name = wdi.loc[wdi['Country Code'] == max_country_code, ['Country Name']].iloc[0,0]
print('The country that has the highest GDP per capita, a Gini coefficient below average for its size classification and a level of human capital below average for its size classification is', max_country_name)
```

```python
#10. 
# we create a dataframe, from wdi with only the indicator for GDP (already found on exercise 2.4 from part 2.)
gdp_data = wdi.loc[wdi['Indicator Code'] == gdp_code].copy()
# then we create a column with the percentage change from 1980 to 2010
gdp_data['Percentage Change'] = ((gdp_data['2010'] - gdp_data['1980']) / gdp_data['1980']) * 100

# then we get the index corresponding to the largest increase (%)
max_change_index_code = gdp_data['Percentage Change'].idxmax(axis = 0)
# finally, having the country code, we search for its name
max_change_country_name = wdi.loc[max_change_index_code, 'Country Name']
print('The country that has the largest % increase in GDP between 1980 and 2010 is', max_change_country_name)
```
